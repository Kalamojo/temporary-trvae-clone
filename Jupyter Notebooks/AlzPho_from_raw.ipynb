```python
import pandas as pd
```

## Loading Data

```python
pho3m = pd.read_csv('./data/alzPro/peaks_proteins_phospho_3m.csv')
pho6m = pd.read_csv('./data/alzPro/peaks_proteins_phospho_6m.csv')
pho9m = pd.read_csv('./data/alzPro/peaks_proteins_phospho_9m.csv')

print(pho3m.columns)
print(pho6m.columns)
print(pho9m.columns)
```

## Column Renaming

```python
pho3m_v2 = pho3m.drop(['Protein Group', 'Protein ID', 'Significance',
                        'Coverage (%)', '#Peptides', '#Unique', 'PTM',
                        'Sample Profile (Ratio)', 'HET Intensity', 
                        'WT Intensity', 'Group Profile (Ratio)', 'Avg. Mass', 
                        'Description'],
                        axis=1)
pho3m_v2 = pho3m_v2.reindex(sorted(pho3m_v2.columns), axis=1)
pho3m_v2.columns
```

```python
pho6m_v2 = pho6m.drop(['Protein Group', 'Protein ID', 'Significance',
                        'Coverage (%)', '#Peptides', '#Unique', 'PTM', 
                        'Sample Profile (Ratio)', 'WT Intensity', 
                        'HET Intensity', 'Group Profile (Ratio)', 'Avg. Mass',
                        'Description'], axis=1)
pho6m_v2 = pho6m_v2.reindex(sorted(pho6m_v2.columns), axis=1)
pho6m_v2.columns
```

```python
pho9m_v2 = pho9m.drop(['Protein Group', 'Protein ID', 'Significance',
                        'Coverage (%)', '#Peptides', '#Unique', 'PTM', 
                        'Sample Profile (Ratio)', 'WT Intensity', 
                        'HET Intensity', 'Group Profile (Ratio)', 'Avg. Mass', 
                        'Description'], axis=1)
pho9m_v2.rename(columns={'Accession': 'Accession Intensity'}, inplace=True)
pho9m_v2.rename(columns=lambda x: x[:-10], inplace=True)
pho9m_v2 = pho9m_v2.reindex(sorted(pho9m_v2.columns), axis=1)
pho9m_v2.columns
```

## Data Combination

### Removing Exclusive Proteins

```python
pho3m_v2["Accession"] = pho3m_v2["Accession"].apply(lambda x: x[:-6])
pho6m_v2["Accession"] = pho6m_v2["Accession"].apply(lambda x: x[:-6])
pho9m_v2["Accession"] = pho9m_v2["Accession"].apply(lambda x: x[:-6])
pho3m_v2.head()
```

```python
phosphors_list_df = pho3m_v2.loc[pho3m_v2["Accession"].isin(pho6m_v2["Accession"])]
phosphors_list = phosphors_list_df.loc[phosphors_list_df["Accession"].isin(pho9m_v2["Accession"])]["Accession"]
len(phosphors_list)
```

```python
phosphors_list
```

```python
pho3m_v3 = pho3m_v2.loc[pho3m_v2["Accession"].isin(phosphors_list)]
pho6m_v3 = pho6m_v2.loc[pho6m_v2["Accession"].isin(phosphors_list)]
pho9m_v3 = pho9m_v2.loc[pho9m_v2["Accession"].isin(phosphors_list)]
```

#### V3 Phosphorylation Merging

```python
pho3m_v3.sort_values(by=["Accession"], inplace=True)
pho6m_v3.sort_values(by=["Accession"], inplace=True)
pho9m_v3.sort_values(by=["Accession"], inplace=True)

pho3m_v3.reset_index(drop=True, inplace=True)
pho6m_v3.reset_index(drop=True, inplace=True)
pho9m_v3.reset_index(drop=True, inplace=True)

pho3m_v3.head()
```

```python
pho6m_v3.head()
```

```python
pho9m_v3.head()
```

```python
phosphors_v3 = pd.concat([pho3m_v3, pho6m_v3.drop(columns=["Accession"], axis=1), pho9m_v3.drop(columns=["Accession"], axis=1)], axis=1)
phosphors_v3.rename(columns=lambda x: x + "_Pho", inplace=True)
phosphors_v3.rename(columns={'Accession_Pho': 'Accession'}, inplace=True)
phosphors_v3
```

#### V3 Cleaning and Saving

```python
phosphors_v3.index = phosphors_v3["Accession"]
phosphors_v3.drop(columns=["Accession"], inplace=True)
phosphors_v3
```

```python
proObs = pd.read_excel("./data/alzPro/alz_sample_id.xlsx")

proObs
```

```python
t1 = phosphors_v3.T
t1.reset_index(drop=True, inplace=True)
t2 = proObs
t3 = pd.DataFrame(index=phosphors_v3.index)
t3.head()
```

```python
adata = sc.AnnData(X=t1, obs=t2, var=t3)
```

```python
print(adata)
```

```python
adata.write_h5ad("./data/alzPho_count.h5ad")
```

### Load Proteins

```python
proteins_v2 = pd.read_csv("./data/alzPro/proteins_v2.csv")
proteins_v2.head()
```

```python
phosphors_list = proteins_v2.loc[proteins_v2["Accession"].isin(phosphors_list)]["Accession"]
print(phosphors_list)
```

```python
pho3m_v2 = pho3m_v2.loc[pho3m_v2["Accession"].isin(phosphors_list)]
pho6m_v2 = pho6m_v2.loc[pho6m_v2["Accession"].isin(phosphors_list)]
pho9m_v2 = pho9m_v2.loc[pho9m_v2["Accession"].isin(phosphors_list)]
proteins_v2 = proteins_v2.loc[proteins_v2["Accession"].isin(phosphors_list)]
```

### Actual Merging

```python
pho3m_v2.sort_values(by=["Accession"], inplace=True)
pho6m_v2.sort_values(by=["Accession"], inplace=True)
pho9m_v2.sort_values(by=["Accession"], inplace=True)
proteins_v2.sort_values(by=["Accession"], inplace=True)

pho3m_v2.reset_index(drop=True, inplace=True)
pho6m_v2.reset_index(drop=True, inplace=True)
pho9m_v2.reset_index(drop=True, inplace=True)
proteins_v2.reset_index(drop=True, inplace=True)

pho3m_v2.head()
```

```python
pho6m_v2.head()
```

```python
pho9m_v2.head()
```

```python
proteins_v2
```

```python
phosphors_v2 = pd.concat([pho3m_v2, pho6m_v2.drop(columns=["Accession"], axis=1), pho9m_v2.drop(columns=["Accession"], axis=1)], axis=1)
phosphors_v2.rename(columns=lambda x: x + "_Pho", inplace=True)
phosphors_v2.rename(columns={'Accession_Pho': 'Accession'}, inplace=True)
phosphors_v2
```

```python
peaks_proteins = pd.concat([proteins_v2, phosphors_v2.drop(columns=["Accession"], axis=1)], axis=1)
peaks_proteins
```

## Cleaning and Saving

```python
peaks_proteins.index = peaks_proteins["Accession"]
peaks_proteins.drop(columns=["Accession"], inplace=True)
peaks_proteins
```

```python
proObs = pd.read_excel("./data/alzPro/alz_sample_id_Pho.xlsx")

proObs
```

```python
t1 = peaks_proteins.T
t1.reset_index(drop=True, inplace=True)
t2 = proObs
t3 = pd.DataFrame(index=peaks_proteins.index)
t3.head()
```

```python
adata = sc.AnnData(X=t1, obs=t2, var=t3)
```
